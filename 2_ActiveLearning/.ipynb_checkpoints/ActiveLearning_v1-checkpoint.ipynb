```python
def get_seed(y_train,size=10):
    # problem is that the initial seed values must include at least one value from each cases
    # find the first 1 and 0 in the dataset
    df2=pd.DataFrame(y_train)
    df2.columns=['match']
    first0=df2[df2['match']==0].index[0] # first apperance of 0
    first1=df2[df2['match']==1].index[0] # first apperance of 1
    # if 1 appeared later 0 in the dataset
    if (first0<first1):
        nlabelbeg=first1-size
        nlabelend=first1+1
    # if 1 appeared before 0 in the dataset    
    else:
        nlabelbeg=first0-size
        nlabelend=first0+1
    
    return nlabelbeg, nlabelend
```

```python
# inspired by http://billchambers.me/tutorials/2014/12/22/cosine-similarity-explained-in-python.html
import string
import math
def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)
```

```python
i=0
tokenize = lambda doc: doc.lower().split(" ")
while(i<len(jac_columns)):
    ls=[]
    k=0
    list_page=list(p_result_nomiss[page_columns[i]])
    list_ca=list(p_result_nomiss[cat_columns[i]])
    tokenized_page = [tokenize(d) for d in list_page] # tokenized docs
    tokenized_ca = [tokenize(d) for d in list_ca] # tokenized docs
    while(k<p_result_nomiss.shape[0]):
        ls.append(jaccard_similarity(tokenized_page[k],tokenized_ca[k]))
        k=k+1
    p_result_nomiss[jac_columns[i]]=ls
    i=i+1
```

```python
#!/usr/bin/env python3
"""
The script helps guide the users to quickly understand how to use
libact by going through a simple active learning task with clear
descriptions.
"""

import copy
import os

import numpy as np
import matplotlib.pyplot as plt
try:
    from sklearn.model_selection import train_test_split
except ImportError:
    from sklearn.cross_validation import train_test_split

# libact classes
from libact.base.dataset import Dataset, import_libsvm_sparse
from libact.models import *
from libact.query_strategies import *
from libact.labelers import IdealLabeler


def run(trn_ds, tst_ds, lbr, model, qs, quota):
    E_in, E_out = [], []

    for _ in range(quota):
        # Standard usage of libact objects
        # make_query returns the index of the sample that the active learning algorithm would like to query
        ask_id = qs.make_query()
        X, _ = zip(*trn_ds.data)
        # lbl label returns the label of a given sample
        lb = lbr.label(X[ask_id])
        # update updates the unlabeled sample with queried sample
        trn_ds.update(ask_id, lb)

        model.train(trn_ds)
        E_in = np.append(E_in, 1 - model.score(trn_ds))
        E_out = np.append(E_out, 1 - model.score(tst_ds))

    return E_in, E_out


def split_train_test(dataset_filepath, test_size, n_labeled):
    
    X, y = import_libsvm_sparse(dataset_filepath).format_sklearn()

    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=test_size)
    trn_ds = Dataset(X_train, np.concatenate([y_train[:n_labeled], [None] * (len(y_train) - n_labeled)]))
    tst_ds = Dataset(X_test, y_test)
    fully_labeled_trn_ds = Dataset(X_train, y_train)

    return trn_ds, tst_ds, y_train, fully_labeled_trn_ds
    
```

```python
import copy
import os

import numpy as np
import matplotlib.pyplot as plt
try:
    from sklearn.model_selection import train_test_split
except ImportError:
    from sklearn.cross_validation import train_test_split

# libact classes
from libact.base.dataset import Dataset, import_libsvm_sparse
from libact.models import *
from libact.query_strategies import *
from libact.labelers import IdealLabeler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score


def run(trn_ds, tst_ds, lbr, model, qs, quota, X_test, y_test):
    # the main active learning algorithm
    E_in, E_out = [], []
    f1score=[]
    tn, fp, fn, tp =[], [], [], []
    
    i = 0
    for _ in range(quota):
        # Standard usage of libact objects
        # make_query returns the index of the sample that the active learning algorithm would like to query
        # ask id only asks for particular id, not all, everytime
        ask_id = qs.make_query()
        X, _ = zip(*trn_ds.data)
        # lbl label returns the label of a given sample
        lb = lbr.label(X[ask_id])
        # update updates the unlabeled sample with queried sample
        trn_ds.update(ask_id, lb)
        
        # trains only on the labeled examples
        model.train(trn_ds)
        pred_y=model.predict(X_test)
        f1score.append(f1_score(y_test, pred_y))
        tn.append(confusion_matrix(y_test,pred_y)[0][0])
        fp.append(confusion_matrix(y_test,pred_y)[0][1])
        fn.append(confusion_matrix(y_test,pred_y)[1][0])
        tp.append(confusion_matrix(y_test,pred_y)[1][1])
        
        i=i+1
        if(len(np.unique(np.array(tp)))>1):
            break

        # score returns the mean accuracy of the results
        E_in = np.append(E_in, 1 - model.score(trn_ds)) #train
        E_out = np.append(E_out, 1 - model.score(tst_ds)) #test
        

    return E_in, E_out, f1score, tn, fp,fn, tp, i, trn_ds
```

```python
# number of samples to query
seed=10
quota = len(y_train) - seed
quota=300
```
