- calculate similarities

```python
page_columns=['page_brand','page_phone_type','page_memory']
cat_columns=['cat_brand','cat_phone_type','cat_memory']
jac_columns=['jac_brand','jac_phone_type','jac_memory']
```

```python
# inspired by http://billchambers.me/tutorials/2014/12/22/cosine-similarity-explained-in-python.html
import string
import math
def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)
```

```python
i=0
tokenize = lambda doc: doc.lower().split(" ")
while(i<len(jac_columns)):
    ls=[]
    k=0
    list_page=list(p_result_nomiss[page_columns[i]])
    list_ca=list(p_result_nomiss[cat_columns[i]])
    tokenized_page = [tokenize(d) for d in list_page] # tokenized docs
    tokenized_ca = [tokenize(d) for d in list_ca] # tokenized docs
    while(k<p_result_nomiss.shape[0]):
        ls.append(jaccard_similarity(tokenized_page[k],tokenized_ca[k]))
        k=k+1
    p_result_nomiss[jac_columns[i]]=ls
    i=i+1
```

```python
#!/usr/bin/env python3
"""
The script helps guide the users to quickly understand how to use
libact by going through a simple active learning task with clear
descriptions.
"""

import copy
import os

import numpy as np
import matplotlib.pyplot as plt
try:
    from sklearn.model_selection import train_test_split
except ImportError:
    from sklearn.cross_validation import train_test_split

# libact classes
from libact.base.dataset import Dataset, import_libsvm_sparse
from libact.models import *
from libact.query_strategies import *
from libact.labelers import IdealLabeler


def run(trn_ds, tst_ds, lbr, model, qs, quota):
    E_in, E_out = [], []

    for _ in range(quota):
        # Standard usage of libact objects
        # make_query returns the index of the sample that the active learning algorithm would like to query
        ask_id = qs.make_query()
        X, _ = zip(*trn_ds.data)
        # lbl label returns the label of a given sample
        lb = lbr.label(X[ask_id])
        # update updates the unlabeled sample with queried sample
        trn_ds.update(ask_id, lb)

        model.train(trn_ds)
        E_in = np.append(E_in, 1 - model.score(trn_ds))
        E_out = np.append(E_out, 1 - model.score(tst_ds))

    return E_in, E_out


def split_train_test(dataset_filepath, test_size, n_labeled):
    
    X, y = import_libsvm_sparse(dataset_filepath).format_sklearn()

    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=test_size)
    trn_ds = Dataset(X_train, np.concatenate([y_train[:n_labeled], [None] * (len(y_train) - n_labeled)]))
    tst_ds = Dataset(X_test, y_test)
    fully_labeled_trn_ds = Dataset(X_train, y_train)

    return trn_ds, tst_ds, y_train, fully_labeled_trn_ds
    
```

### 1. Logistic Regression

```python
import copy
import os

import numpy as np
import matplotlib.pyplot as plt
try:
    from sklearn.model_selection import train_test_split
except ImportError:
    from sklearn.cross_validation import train_test_split

# libact classes
from libact.base.dataset import Dataset, import_libsvm_sparse
from libact.models import *
from libact.query_strategies import *
from libact.labelers import IdealLabeler

def run(trn_ds, tst_ds, lbr, model, qs, quota):
    # the main active learning algorithm
    E_in, E_out = [], []

    for _ in range(quota):
        # Standard usage of libact objects
        # make_query returns the index of the sample that the active learning algorithm would like to query
        ask_id = qs.make_query()
        X, _ = zip(*trn_ds.data)
        # lbl label returns the label of a given sample
        lb = lbr.label(X[ask_id])
        # update updates the unlabeled sample with queried sample
        trn_ds.update(ask_id, lb)

        model.train(trn_ds)
        E_in = np.append(E_in, 1 - model.score(trn_ds)) #train
        E_out = np.append(E_out, 1 - model.score(tst_ds)) #test

    return E_in, E_out
```

```python
# split the dataset
p_result_nomiss_jac=p_result_nomiss[['jac_brand','jac_phone_type','match','jac_memory']]
X_train, X_test, y_train, y_test = train_test_split(p_result_nomiss_jac.drop('match',axis=1).values,p_result_nomiss_jac['match'].values,test_size=0.33,random_state=42)
# the data needs to be in Dataset class
# we have 10 labeled instances, make the other instances 'None'
trn_ds = Dataset(X_train, np.concatenate([y_train[:n_labeled], [None] * (len(y_train) - n_labeled)]))
trn_ds2 = copy.deepcopy(trn_ds)
tst_ds = Dataset(X_test, y_test)
fully_labeled_trn_ds = Dataset(X_train, y_train)
lbr = IdealLabeler(fully_labeled_trn_ds)
```

```python
n_labeled=300
# number of samples to query
quota = len(y_train) - n_labeled
```

```python
# uncertainity sampling
qs = UncertaintySampling(trn_ds, method='lc', model=LogisticRegression())
model = LogisticRegression()
quota=200
E_in_1, E_out_1 = run(trn_ds, tst_ds, lbr, model, qs, quota)
```

```python
# random sampling
qs2 = RandomSampling(trn_ds2)
model = LogisticRegression()
E_in_2, E_out_2 = run(trn_ds2, tst_ds, lbr, model, qs2, quota)
```

```python
    # Plot the learning curve of UncertaintySampling to RandomSampling
    # The x-axis is the number of queries, and the y-axis is the corresponding
    # error rate.
    query_num = np.arange(1, quota + 1)
    plt.plot(query_num, E_in_1, 'b', label='qs Ein')
    plt.plot(query_num, E_in_2, 'r', label='random Ein')
    plt.plot(query_num, E_out_1, 'g', label='qs Eout')
    plt.plot(query_num, E_out_2, 'k', label='random Eout')
    plt.xlabel('Number of Queries')
    plt.ylabel('Error')
    plt.title('Experiment Result')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
               fancybox=True, shadow=True, ncol=5)
    plt.show()
```
