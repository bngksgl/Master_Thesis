```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_dataset(path, k='/Results/'):
    # qbc
    df=pd.read_csv(os.getcwd()+k+path)

    total1=[]
    for i in df['label']:
        a=i.replace('[','').replace(']','').split(',')
        total1.append(sum([int(k.strip()) for k in a]))
    df['total1s']=total1

    sm=0
    ls=[]
    for i in df['total1s']:
        sm=sm+i
        ls.append(sm)

    df['total1s_cum']=ls

    return df
```

- Paper

```python
f1scores=pd.read_excel('HeadPhone_Results.xlsx','Phones_forthesis')
f1scores.head()
```

```python
f1scores_main=pd.read_excel('HeadPhone_Results.xlsx','all_trained')
f1scores_main.head()
```

```python
i=0
iter_=[]
quota=[]
f1_score=[]
while(i<f1scores_main.shape[0]):
    sub_df=pd.read_csv(os.getcwd()+'/Results_changename/'+f1scores_main['full_name'].loc[i])
    sub_df['f1_score']=[round(x,2) for x in sub_df['f1_score']]
    
    mis=f1scores_main['full_name'].loc[i].split('_')[2]
    fet=f1scores_main['full_name'].loc[i].split('_')[3]
    alg=f1scores_main['full_name'].loc[i].split('_')[5]
    
    f1=list(f1scores[(f1scores['Variables']==fet)&(f1scores['Missing option']==mis)&(f1scores['Algorithm']==alg)].values)[0][-1]
    f1_score.append(f1)
    
    iter_.append(sub_df[sub_df['f1_score']>=f1]['iter'].values[0])
    quota.append(sub_df[sub_df['f1_score']>=f1]['quota'].values[0])
    i=i+1
```

```python
f1scores_main['iteration']=iter_
f1scores_main['quota']=quota
f1scores_main['F1 Score']=f1_score
```
